```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df=pd.read_csv(r"C:\Users\Megha\Desktop\School Completion Rates.csv")
```

```python
#display the top 5 rows of dataset
df.head()
```

```python
#display the rows and column numbers of the dataset in tuple format
df.shape
```

```python
#remove the features with no feature name
df.drop(labels=["Unnamed: 18","Unnamed: 19","Unnamed: 20","Unnamed: 21","Unnamed: 22","Unnamed: 23"],axis=1,inplace=True)
```

```python
df.columns
```

```python
#display the basic information & overview of the dataset
df.info()
```

```python
#find how many null values are present
df.isna().sum()
```

```python
#remove the features with too many null values
df.drop(labels=["Least developed countries (LDC)","Africa sub-regions","Africa region","Sub-region","Source"]
        ,axis=1,inplace=True)
```

```python
df["Total"].fillna(df["Total"].mean(),inplace=True)
df["Male"].fillna(df["Male"].mean(),inplace=True)
df["Female"].fillna(df["Female"].mean(),inplace=True)
df["Urban"].fillna(df["Urban"].mean(),inplace=True)
df["Rural"].fillna(df["Rural"].mean(),inplace=True)
df["Poorest"].fillna(df["Poorest"].mean(),inplace=True)
df["Second"].fillna(df["Second"].mean(),inplace=True)
df["Middle"].fillna(df["Middle"].mean(),inplace=True)
df["Fourth"].fillna(df["Fourth"].mean(),inplace=True)
df["Richest"].fillna(df["Richest"].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df["Region"].unique()
```

```python
sns.countplot(x=df["Region"],data=df,color="magenta")
```

```python
df["Region"].fillna("ECA",inplace=True)
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
sns.histplot(x=df["Total"],color="red")

plt.subplot(2,2,2)
sns.histplot(x=df["Male"],color="blue")

plt.subplot(2,2,3)
sns.histplot(x=df["Female"],color="yellow")

plt.subplot(2,2,4)
sns.histplot(x=df["Middle"],color="green")
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
sns.scatterplot(x="Urban",y="Rural",data=df,hue="Region",marker="*")
sns.rugplot(x="Urban",y="Rural",data=df,hue="Region",height=0.1)

plt.subplot(2,2,2)
sns.scatterplot(x="Poorest",y="Second",data=df,hue="Region",marker="*")
sns.rugplot(x="Poorest",y="Second",data=df,hue="Region",height=0.1)

plt.subplot(2,2,3)
sns.scatterplot(x="Middle",y="Fourth",data=df,hue="Region",marker="*")
sns.rugplot(x="Middle",y="Fourth",data=df,hue="Region",height=0.1)

plt.subplot(2,2,4)
sns.scatterplot(x="Total",y="Richest",data=df,hue="Region",marker="*")
sns.rugplot(x="Total",y="Richest",data=df,hue="Region",height=0.1)
```
