```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#load the default datasets
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

## Distribution Plots

```python
#the line is called Kernel Density Estimation(KDE)
sns.distplot(x=tips["total_bill"])
plt.title("Kernel Density Estimation")
```

```python
#removing KDE
sns.distplot(tips["total_bill"],color="violet",kde=False)
plt.title("Kernel Density Estimation")
```

```python
sns.kdeplot(tips["total_bill"],color="violet")
```

```python
#bins=>the number of boxes of histogram
sns.distplot(a=tips["total_bill"],color="violet",kde=False,bins=50)
```

## Categorical Plots

```python
#parameters are x values, y value, variable in which dataset is loaded into
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex",color="crimson")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde",hue="smoker")
```

```python
#show variety of plots and if dataset is huge then it takes longer time for execution
#hue=>separate colours for different categories in column
#on using hue automatically legend is generated
sns.pairplot(tips,hue="time",markers=["s","D"],corner=True)
```

```python
#when only x or y is given then we dashes only for 1 axis
#draws dash marks for every point
#number of dash represent how much distribution is present
sns.rugplot(x=tips["total_bill"],y=tips["tip"],color="darkorange",height=0.2,hue=tips["time"])
```

```python
#can execute only 1 variable at a time
#histogram bin plot draws bins for each point 
#height of bin represent how much distribution is present
sns.histplot(tips["total_bill"],color="pink",edgecolor="black")
```

```python
#can compare categorical variable with numerical variable and gives a brief relationship between both
sns.barplot(x="smoker",y="total_bill",data=tips,color="purple")
```

```python
sns.histplot(x="day",data=tips,color="pink",edgecolor="blue")
sns.countplot(x="day",data=tips,color="purple",alpha=0.5)
```

```python
#marker=>shape of the representation of datapoint
#hue=>color differentiation on basis of this parameter
sns.scatterplot(x="smoker",y="total_bill",data=tips,marker="+",color="maroon",hue="sex")
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,color="maroon",hue="smoker",marker="+")
sns.rugplot(x="tip",y="total_bill",data=tips,color="darkorange",hue="smoker",height=0.1)
```

```python
#takes only one variable as argument and gives the respective counts of each category
sns.countplot(x="sex",data=tips,color="blue",hue="smoker")
```

```python
#gives values like=>min,lower quartile,median,upper quartile,max
#distance between min & lower quartile, upper quartile & max=>whisker
#distance between lower quartile and upper quartile=>Interquartile Range
#hue(optional)=>another category that divides the box plot on whether a person is smoker or not
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker",color="gray")
```

```python
#similar to box plot but has 1 extra parameter that is bulging
#more bulging=>more data points
#hue(optional)=>another category that divides the box plot on whether a person is smoker or not
#split(optional)=>makes the two plots joined and this makes the plot professional and preferable
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",split=True,color="cyan")
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,color="gray",hue="smoker")
sns.violinplot(x="day",y="total_bill",data=tips,color="cyan",hue="smoker",split=True)
```

```python
#similar to scatter plot as well as violin plot
#more dense means more datapoints
#hue(optional)=>another category that divides the box plot on whether a person is smoker or not
sns.stripplot(x="day",y="total_bill",data=tips,hue="smoker",marker="*",color="olive")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",split=True,color="cyan")
sns.stripplot(x="day",y="total_bill",data=tips,hue="smoker",marker="*",color="olive")
```

## Matrix Plots

```python
#loading the dataset
flights=sns.load_dataset("flights")
```

```python
flights.head()
```

```python
#dataset should have some relation between row name and column name
#like if row name=>student 1 & student 2, column name=>test 1 & test 2
#there is relation between row name & column name
#we will convert this to a pivot table as there is no relation between row name and column name
fp=flights.pivot_table(index="month",columns="year",values="passengers")
fp
```

```python
#dataset should be in form of matrix which means dataframe to draw heatmaps
#cmap(optional)=>color of maps,linewidth(optional)=>space between each small box,linecolor=>color between each small box
sns.heatmap(fp,cmap="magma",linewidth=1,linecolor="black")
```

```python
sns.heatmap(fp,cmap="coolwarm",linewidth=1,linecolor="magenta",vmin=100,vmax=500,center=250)
```

```python
#groups similar data together which have similar color scheme
#cmap(optional)=>color of maps
#"coolwarm" is a color scheme which is used by maximum data scientists
sns.clustermap(fp,cmap="coolwarm")
```

## Grids

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris["species"].unique()
```

```python
#we are not in control of what is being printed in pair plot
sns.pairplot(iris)
```

```python
pg=sns.PairGrid(iris)
#diagonals are distribution plots
pg.map_diag(sns.distplot)
#upper plots of diagonals are scatter plots
pg.map_upper(sns.scatterplot)
#lower plots of diagonals are kernel density estimation(KDE) plots
pg.map_lower(sns.kdeplot)
```

```python
pg=sns.PairGrid(iris,hue="species")
pg.map_diag(sns.distplot)
pg.map_upper(sns.scatterplot)
pg.map_lower(sns.kdeplot)
```

```python
pg.savefig("Plot Image.jpg",dpi=200)
```
