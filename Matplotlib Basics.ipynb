```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#helps to show output of plots
#no need to use plt.show() command
#types of plotting=functional approach, object-oriented approach
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

## Function Approach

```python
plt.plot(x,y,"m")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Graph")
```

```python
#plotting multiple plots
#parameters of subplot=no of rows,no of columns, plot number
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"y")
```

## Object Oriented Approach

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Axis")
axes.set_ylabel("Y Axis")
axes.set_title("Simple Graph")
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y)
axes2.set_title("Smaller Plot")
```

## Saving A Plot


```python
#dpi=dots per image
#saves plot where jupyter is stored
fig.savefig("Plot Image.jpg",dpi=200)
```

## Adding Legend To Plots 

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="X Squared Value")
ax.plot(x,x**3,label="X Cubed Value")

#legend()=>does not take any parameters but identifies graphs on labels
#loc=parameter that decides location of legend box
ax.legend(loc="upper right")
```

## Customising plots

```python
#alpha=>parameter for transparency, color=>parameter for colours
#lw=>linewidth, ls=>linespace
#marker=>plots the values of y vs x on graph
plt.plot(x,y,color="m",alpha=0.7,lw=4,ls=":",marker="h")
```
